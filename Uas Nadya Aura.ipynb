```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Baca gambar
image = cv2.imread('aura2.jpg')
```

```python
# Konversi gambar ke grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Terapkan Gaussian Blur untuk mengurangi noise
blurred = cv2.GaussianBlur(gray, (5, 5), 1.4)
```

```python
# Terapkan Canny Edge Detection
edges = cv2.Canny(blurred, 100, 200)
```

```python
# Deteksi kontur
contours, hierarchy = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# Gambar kontur pada gambar asli
contour_image = image.copy()
cv2.drawContours(contour_image, contours, -10, (0, 255, 0), 5)
```

```python
# Output Pertama: Gambar Asli dan Hasil Deteksi Tepi dengan Canny
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.title('Canny Edge Detection')
plt.imshow(edges, cmap='gray')

plt.show()
```

```python
# Output Kedua: Gambar Asli, Hasil Deteksi Tepi dengan Canny, dan Hasil Deteksi Kontur
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

plt.subplot(1, 3, 2)
plt.title('Canny Edge Detection')
plt.imshow(edges, cmap='gray')

plt.subplot(1, 3, 3)
plt.title('Contours Detection')
plt.imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))

plt.show()
```
